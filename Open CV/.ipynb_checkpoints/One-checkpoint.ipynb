### Blurring an Image

```python
blur_img = cv2.GaussianBlur(gray_img,(9,9),0) # kernel should be odd number, sigma->0
plt.imshow(blur_img, cmap='gray')
```

### Edge Detection

```python
def canny_edge_detection(img):
    return cv2.Canny(img,100,100)
```

```python
canny_img = canny_edge_detection(gray_img)
plt.imshow(canny_img, cmap="gray");plt.axis("off")
```

### Image Dilation

```python
kernel = np.ones((5,5), np.uint8)
dilated_img = cv2.dilate(canny_img, kernel, iterations=1) # Iteration/Thickness = 1
plt.imshow(dilated_img, cmap="gray");plt.axis("off")
```

### Image Erosion

```python
kernel = np.ones((5,5), np.uint8)
eroded_img = cv2.erode(dilated_img, kernel, iterations=1)
plt.imshow(eroded_img, cmap="gray");plt.axis("off")
```

### BITWISE Operations

```python
blank = np.zeros((400,400), dtype='uint8')
```

```python
rectangle = cv2.rectangle(blank.copy(), (30,30), (370,370), 255, -1) # -1=thickness
circle = cv2.circle(blank.copy(), (200,200), 200, 255, -1) # 200=radius
```

```python
plt.imshow(rectangle, cmap="gray")
```

```python
plt.imshow(circle, cmap="gray")
```

```python
# Bitwise And
bitwise_and = cv2.bitwise_and(rectangle,circle)
plt.imshow(bitwise_and, cmap="gray")
```

```python
# Bitwise or
bitwise_or = cv2.bitwise_or(rectangle,circle)
plt.imshow(bitwise_or, cmap="gray")
```

```python
# Bitwise xor
bitwise_xor = cv2.bitwise_xor(rectangle,circle)
plt.imshow(bitwise_xor, cmap="gray")
```

```python
# Bitwise not (Inversion)
bitwise_not = cv2.bitwise_not(rectangle)
plt.imshow(bitwise_not, cmap="gray")
```

# WarpAffine
# -----------------

### Image Translation

```python
def translate(img, x, y):
    """
        -x = LEFT
        +x = RIGHT
        -y = UP
        +y = DOWN
    """
    
    trans_mat = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1],img.shape[0])
    return cv2.warpAffine(img, trans_mat, dimensions)
```

```python
translated = translate(img, -100, 100)
plt.imshow(translated[:,:,::-1])
```

### Image Rotation

```python
def rotate(img, angle, rotPoint=None):
    h, w = img.shape[:2]
    
    if rotPoint == None:
        # Rotate around the center
        rotPoint = (w//2,h//2)
        
    rotMat = cv2.getRotationMatrix2D(rotPoint, angle, 1.0) # 1.0 means we aren't scaling while rotating
    dimensions = (w,h)
    
    return cv2.warpAffine(img, rotMat, dimensions)
    
```

```python
rotated = rotate(img, 40)
plt.imshow(rotated[:,:,::-1])
```

### Flipping

```python
"""
0 = Vertically
1 = Horizontally
-1 = Both
"""
flipped = cv2.flip(img, 0)
plt.imshow(flipped[:,:,::-1])
```
